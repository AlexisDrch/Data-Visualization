```python
import pandas as pd
```

```python
df = pd.read_csv('all-ages.csv')
df_less = df.loc[:,['Major_category', 'Major', 'Employed', 'Unemployed', 'Unemployment_rate']]
df_grouped = df_less.groupby(['Major_category', 'Major'], sort=True).mean()
```

```python
def color_max_white(x):
    
    return 'color: white'

def highlight_max(x): 
    c1 = 'background-color: red'
    c2 = '' 
    mask_max = x.groupby(level=0)['Unemployment_rate'].transform('max').eq(x['Unemployment_rate'])
    
    df1 = pd.DataFrame(c2, index=x.index, columns=x.columns)
    df1.loc[mask_max, ['Employed', 'Unemployed', 'Unemployment_rate']] = c1
    return df1

df_grouped.style.apply(highlight_max,axis=None)
```

```python
df_grouped.style.format({'Unemployment_rate': "{:.2%}"})
```

```python
#export to html
df_grouped.to_html('table_all_ages.html', bold_rows=False,)
```
